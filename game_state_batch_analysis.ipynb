```python
from poker.player_deep_agent import PlayerDeepAgent
```

```python
file = 'game_state_batches_PlayerRandom_2_PlayerDeepAgent_2_1000.pkl' 
with open(file, 'rb') as f:
    game_state_batches = pickle.load(f)
```

```python
# now analyze the game states of a single game from the perspective of a Deep Agent Player
game_state_batch_1 = game_state_batches[0]
player = list(game_state_batch_1.keys())[0]

game_states = game_state_batch_1[player]

for round, game_state in enumerate(game_states):
    pre_stack, post_stack, stack_delta, gs = game_state
    #game_state_, action_probs, raise_ratio, action = gs
    print(f"Round: {round}")
    print(f"Pre-Round Stack: {pre_stack}")
    print(f"Post-Round Stack: {post_stack}")
    print(f"Stack delta: {stack_delta:.4f}")

    print("Game states and actions within round:\n")
    for game_states_round in gs:
        game_state_, action_probs, raise_ratio, action = game_states_round

        print(f"\nState:\n{game_state_}")
        print(f"Action probs: {action_probs}")
        print(f"Raise Ratio: {raise_ratio:.4f}")
        print(f"Action Taken: {action}")
    
    print(f"{'-'*60}\n\n")
        
```
