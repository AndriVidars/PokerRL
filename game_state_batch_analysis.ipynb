```python
import pickle
import numpy as np
```

```python
def analyze_deep_player_actions(game_state_file, num_games=1):
    with open(game_state_file, 'rb') as f:
        game_state_batches = pickle.load(f)
    
    game_state_batches = game_state_batches[:num_games]
    for i, gsb in enumerate(game_state_batches):
        print(f"{'-'*80}\nGame: {i+1}\n{'-'*80}\n")
        player = list(gsb.keys())[0]
        game_states = gsb[player]
        for round, game_state in enumerate(game_states):
            pre_stack, post_stack, stack_delta, gs = game_state
            print(f"Round: {round}")
            print(f"Pre-Round Stack: {pre_stack}")
            print(f"Post-Round Stack: {post_stack}")
            print(f"Stack delta: {stack_delta:.4f}")

            print("Game states and actions within round:\n")
            for game_states_round in gs:
                game_state_, action_probs, raise_ratio, action = game_states_round

                print(f"\nState:\n{game_state_}")
                print(f"Action probs: {np.array2string(action_probs, precision=4, floatmode='fixed', suppress_small=True)}")
                print(f"Raise Ratio: {raise_ratio:.4f}")
                print(f"Action Taken: {action}")

            print(f"{'-'*60}\n\n")
```

```python
analyze_deep_player_actions('pkl/game_state_batches_PlayerHeuristic_2_PlayerDeepAgent_2_PlayerPPO_2_1000.pkl', num_games=3) # NOTE output is truncated in nb, can see in text editor
```
